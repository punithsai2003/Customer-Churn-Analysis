```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/customer_churn_dataset-training-master.csv")
data.head()
```

# Data Preprocessing

```python
data.info()
```

```python
data["CustomerID"] = data["CustomerID"].astype(int)
data["Churn"] = data["Churn"].astype(int)
```

```python
data.head()
```

```python
data.info()
```

# Exploratory Data Analysis

```python
#churn rate  by Age Group
data["AgeGroup"] = pd.cut(data["Age"], bins=[0,20,40,60,100], labels = ["<20","20-40","40-60","60+"])
sns.barplot(x="AgeGroup", y="Churn", data = data)
plt.title("Churn rate by Age Group")
plt.show()
```

```python
##churn rate  by Gender
sns.barplot(x="Gender", y="Churn", data = data)
plt.title("Churn rate by Gender")
plt.show()
```

```python
#churn rate  by Subscription type
sns.barplot(x="Subscription Type", y="Churn", data = data)
plt.title("Churn rate by Subscription Type")
plt.show()
```

```python
#churn rate  by Contract Length
sns.barplot(x="Contract Length", y="Churn", data = data)
plt.title("Churn rate by Contract Length")
plt.show()
```

```python
#Churn rate by Support calls
sns.boxplot(x="Churn", y="Support Calls", data=data)
plt.title("Support Calls vs Churn")
plt.show()
```

```python
# Correlation Heatmap
corr = data.select_dtypes(include=["float64", "int"]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# **Conclusion**


*   Senior customers (aged 60 and above) exhibited the highest churn rates,
suggesting that age may be a significant factor influencing customer retention.
*   Female customers were found to churn at a slightly higher rate compared to males, highlighting a potential gender-based difference in customer satisfaction or engagement.


*   Customers subscribed to the Basic subscription plan were more likely to churn than those on higher-tier plans, indicating that service value perception may play a role in retention.
*  A clear pattern emerged showing that customers with low or no support interaction (i.e., fewer support calls) had higher churn rates. This may reflect disengagement or unmet service needs.



*   The correlation analysis further supported these findings by identifying strong relationships between churn and factors such as tenure, payment delay, and total spend.
